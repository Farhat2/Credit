```python
from sklearn.ensemble import IsolationForest

# Initialize Isolation Forest with parameters
iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.0017,   # Approximate fraud ratio
    random_state=42
)
```

```python
iso_forest.fit(X_scaled)
```

```python
predictions = iso_forest.predict(X_scaled)
```

```python
y_pred = np.where(predictions == -1, 1, 0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

def plot_conf_mat(Y, y_pred):
    """
    Plots a confusion matrix using Seaborn's heatmap().
    """
    fig, ax = plt.subplots(figsize=(3, 3))
    ax = sns.heatmap(confusion_matrix(Y, y_pred),
                     annot=True, 
                     cbar=False)
    plt.xlabel("True label")
    plt.ylabel("Predicted label")
    
    # Fix the broken annotations (this happened in Matplotlib 3.1.1)
    bottom, top = ax.get_ylim()
    ax.set_ylim(bottom + 0.5, top - 0.5);
    
plot_conf_mat(Y, y_pred)
```

```python
from sklearn.decomposition import PCA

# Initialize PCA to reduce to 2 components
pca = PCA(n_components=2, random_state=42)

# Fit PCA on scaled features and transform
X_pca = pca.fit_transform(X_scaled)

# Check shape of the transformed data
print(X_pca.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], s=1, alpha=0.5)
plt.title('PCA: First two components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.svm import OneClassSVM

# Initialize One-Class SVM with RBF kernel and nu roughly equal to fraud ratio (~0.0017)
oc_svm = OneClassSVM(kernel='rbf', nu=0.0017, gamma='scale')

# Fit the model on PCA data
oc_svm.fit(X_pca)

# Predict anomalies (-1 = anomaly, 1 = normal)
svm_predictions = oc_svm.predict(X_pca)

# Convert predictions to 0 (normal) and 1 (fraud) format
y_svm_pred = np.where(svm_predictions == -1, 1, 0)

print(y_svm_pred[:20])
```

```python
import matplotlib.pyplot as plt

# Metrics for both models
models = ['Isolation Forest', 'One-Class SVM']
precision_scores = [0.258, 0.002]  # Your results
recall_scores = [0.254, 0.748]
f1_scores = [0.256, 0.004]

x = range(len(models))

# Create the bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.25

plt.bar([i - bar_width for i in x], precision_scores, width=bar_width, label='Precision')
plt.bar(x, recall_scores, width=bar_width, label='Recall')
plt.bar([i + bar_width for i in x], f1_scores, width=bar_width, label='F1-Score')

plt.xticks(x, models)
plt.ylabel("Score")
plt.ylim(0, 1)
plt.title("Model Comparison: Isolation Forest vs. One-Class SVM")
plt.legend()
plt.show()
```

📌 Insights
1. Dataset Imbalance
The dataset is highly imbalanced:

Fraudulent transactions (Class 1): 492

Non-fraudulent transactions (Class 0): 284,315

This imbalance explains why accuracy alone is not a reliable metric. We must focus on precision, recall, and F1-score.

2. Isolation Forest Results
Precision: 0.258

Recall: 0.254

F1-score: 0.256

✅ Interpretation:
Isolation Forest provides a good balance between precision and recall. It detects a reasonable portion of fraud cases without producing an excessive number of false positives.

3. One-Class SVM Results
Precision: 0.002

Recall: 0.748

F1-score: 0.004

⚠️ Interpretation:
One-Class SVM achieves very high recall, but its precision is almost zero. This means it labels almost every transaction as fraud, resulting in a large number of false alarms, making it impractical for real-world use.

4. Model Comparison
